# Support Vector Regression (SVR)

## Importing the libraries

```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```
dataset = pd.read_csv('Car_Purchasing_Data.csv', encoding='latin-1')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```
print(X.shape)
print(X)
```

```
print(y)
```

```
y = y.reshape(len(y),1)
```

```
print(y)
```

**Selecting the appropriate features**


```
dataset.describe()
dataset.head()
```

```
print(len(dataset['Country'].unique()))
```

```
X=dataset.iloc[:,3:].values
print(X)
```

**Train Test Split**

```
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test= train_test_split(X,y,test_size=.15)
print(X_test.shape)
```

## Feature Scaling

```
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
sc_y = StandardScaler()
X_train= sc_X.fit_transform(X_train)
y_train = sc_y.fit_transform(y_train)
X_test= sc_X.transform(X_test)
```

## Training the SVR model on the whole dataset

```
from sklearn.svm import SVR
regressor = SVR(kernel = 'rbf')
regressor.fit(X_train, y_train)
ypredd=regressor.predict(X_test)
```

## Predicting a new result

```
ypred=sc_y.inverse_transform(ypredd)
```

```
print(ypred.shape)
ypred.resize(len(ypred),1)

```

```
print(np.concatenate((y_test,ypred),axis=1))
```
